```python
import matplotlib.pyplot as plt
import os
import numpy as np
from celluloid import Camera
from mpl_toolkits.mplot3d import Axes3D
from glob import glob
import matplotlib.animation as animation
```

```python
fig = plt.figure()
ax = Axes3D(fig)
plt.xlim(0,1.0)
plt.ylim(0,1.0)
camera = Camera(fig)
files = sorted([int(x.split("_")[1].split('.')[0]) for x in glob("./points_*.txt")])

for i in files[:]:    
    print (i)
    data = np.genfromtxt("points_{}.txt".format(i), delimiter=',')
#     path = np.genfromtxt("paths_{}.txt".format(i), delimiter=',')
#     ax.scatter(data[:,0], data[:,1], data[:,2], alpha=.2, edgecolor='')
    print ("{:003d}- Percentage of Points' x-value between 0.25 - 0.75: {:.2f}% ".format(
                i, float(np.sum(0.25>data[:,0]) + np.sum(0.75<data[:,0]))/len(data) * 100))
    print ("\t{}".format(float(np.sum(0.25>data[:,0]))/len(data) * 100))
    print ("\t{}".format(float(np.sum(0.75<data[:,0]))/len(data) * 100))
#     if ( len(data[data[:,3]>=1.0]))>0:
#         print (data)
#         break
    ax.scatter(data[:,1], data[:,0], data[:,2], s=data[:,3], color='brown')
#     ax.plot(path[:,1], path[:,0], path[:,2])
#     fig.text(0.5, 0.95, str(i))
    ax.text(1, 0, 0, str(i), color='red')
    ax.set_xlabel("x")
    ax.set_ylabel("y")
#     plt.savefig('points{}.png'.format(i))
#     plt.clf(); plt.cla();plt.close()
    camera.snap()
    
# animation = camera.animate(interval=100)
animation = camera.animate(interval=500)
animation.save('ants_marching.gif', writer = 'imagemagick')
        
```
